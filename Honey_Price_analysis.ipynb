```python
df.drop(columns=["electrical_conduct","glucose_level"],inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,r2_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:8],df.iloc[:,-1],test_size=0.2)
```

```python
le=LabelEncoder()
X_train["pollen"]=le.fit_transform(X_train["pollen"])
X_test["pollen"]=le.transform(X_test["pollen"])
```

```python
X_test
```

```python
df.shape
```

```python
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr=DecisionTreeRegressor(max_depth=None,splitter="random")
dtr.fit(X_train,y_train)
y_pred=dtr.predict(X_test)

print("Train R:", dtr.score(X_train, y_train))
print("Test R :", dtr.score(X_test, y_test))
```
